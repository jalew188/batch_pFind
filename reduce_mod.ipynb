```python
from alphabase.constants.modification import MOD_DF
MOD_DF.loc['Glu->pyro-Glu@E^Any N-term']
```

```python
required_mod = [
    'Oxidation@M',
    'Gln->pyro-Glu@Q^Any N-term',
    'Glu->pyro-Glu@E^Any N-term',
    'Carbamidomethyl@C',
    'Deamidated@N',
]
required_class = [
    'Post-translational',
    'O-linked glycosylation',
    'AA substitution',
    'Multiple',
    #'Non-standard residue',
    'Pre-translational'
]
reduce_df = MOD_DF[
    (
        MOD_DF['classification'].isin(required_class)
        |MOD_DF['name'].isin(required_mod)
    ) & MOD_DF['upper_case_AA'] &
    (MOD_DF['mass'] < 500)
]
reduce_df
```

```python
def to_pfind_ini(df):
    mod_dict = {}
    def get_term_level(site):
        if site.startswith('Protein'):
            return 'PRO'
        else:
            return 'PEP'
    def get_NC_term(site):
        if site.endswith('N-term'):
            return 'N'
        else:
            return 'C'

    def get_term(site):
        return get_term_level(site)+'_'+get_NC_term(site)
    for mod, mass, modloss, composition in df[['name','mass','modloss','composition']].values:
        mod_name, site = mod.split('@')
        if '^' in site:
            mod = f"{mod_name}[{site[2:].replace(' ','')}]"
            AA = site[0]
            site = site[2:]
            if mod in mod_dict:
                mod_dict[mod]['AA'] += AA
            else:
                mod_dict[mod] = {}
                mod_dict[mod]['AA'] = AA
                mod_dict[mod]['info'] = f"{get_term(site)} {mass} {mass} 0 {composition}"
        elif site.endswith('term'):
            mod_name, site = mod.split('@')
            mod = f"{mod_name}[{site.replace(' ','')}]"
            mod_dict[mod] = {}
            mod_dict[mod]['AA'] = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
            mod_dict[mod]['info'] = f"{get_term(site)} {mass} {mass} {'1 '+str(modloss) if modloss else '0'} {composition}"
        else:
            mod_name, site = mod.split('@')
            mod = f"{mod_name}[{site}]"
            mod_dict[mod] = {}
            mod_dict[mod]['AA'] = site
            mod_dict[mod]['info'] = f"{site} {mass} {mass} {'1 '+str(modloss) if modloss else '0'} {composition}"

    with open('reduced_mod.ini','w') as f:
        f.write(f'@NUMBER_MODIFICATION={len(mod_dict)}\n')
        for i,(mod, item) in enumerate(mod_dict.items()):
            AA = item['AA']
            info = item['info']
            if mod.endswith('term]') and len(AA) == 1:
                mod = mod[:-1]+AA+mod[-1]
                print(mod)
            f.write(f'name{i+1}={mod} 0\n')
            f.write(f'{mod}={AA} {info}\n')

to_pfind_ini(reduce_df)
```
